```python
from surfplot import Plot
import pandas as pd 
import numpy as np
import matplotlib.pyplot as plt
import os
from utils import get_annot_labels
from neuromaps.datasets import fetch_fsaverage 
```

```python
base_path = os.path.dirname(os.getcwd())

dat = pd.read_csv(base_path + '/data/STOPPD/parcelwise_groupdifference.csv')
```

```python
# get labels 
lh_labels = get_annot_labels('lh.aparc.annot')    
rh_labels = get_annot_labels('rh.aparc.annot')
```

```python
# make maps to plt
lh_dat = dat.loc[dat['parcel'].str.contains('lh_'),:]
lh_dat.index = range(1, len(lh_dat) + 1)
lh_dict = lh_dat['p-value'].to_dict()
lh_map = np.array([lh_dict.get(index) for index in lh_labels], dtype=np.float64)

rh_dat = dat.loc[dat['parcel'].str.contains('rh_'),:]
rh_dat.index = range(1, len(rh_dat) + 1)
rh_dict = rh_dat['p-value'].to_dict()
rh_map = np.array([rh_dict.get(index) for index in rh_labels], dtype=np.float64)

map = {'left': lh_map, 'right': rh_map}
```

```python
# plot 
surface = fetch_fsaverage(density='164k')
lh, rh = surface['inflated']

p = Plot(lh, rh, views=['lateral', 'medial'], zoom=1.2)
p.add_layer(map, cmap='Reds_r', cbar_label='p-value')
pic = p.build()
```
